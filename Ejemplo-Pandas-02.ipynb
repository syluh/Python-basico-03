```python
# Generar un archivo .csv que contenga únicamente las características:
# Nombre_Institucion
# Provincia 
# Canton
# Parroaquia
# Número de estudiantes
# Número de docentes

# Siempre que cumpla con la condición: el número de estudiantes debe ser mayor a 20 y menor o igual a 100
# El archivo debe tener el nombre: instituciones_0001.csv
```

```python
datafinal1 = datos[['Nombre_Institucion','Provincia','Canton','Parroquia','Numero_de_estudiantes','Numero_de_docentes']]
```

```python
datafinal2 = datafinal1[(datafinal1['Numero_de_estudiantes'] > 20) & (datafinal1['Numero_de_estudiantes'] < 100)]
```

```python
datafinal2.to_csv("data2/instituciones_0001.csv")
```

```python
# Generar un archivo .csv que contenga únicamente las características:
# Nombre_Institucion
# Provincia 
# Canton
# Parroaquia
# Número de estudiantes
# Número de docentes

# Siempre que cumpla con la condición: el número de docentes debe ser mayor a 1 y menor o igual a 50
# El archivo debe tener el nombre: instituciones_0002.csv
```
